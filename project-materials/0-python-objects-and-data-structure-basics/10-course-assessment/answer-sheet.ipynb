 **Numbers:** Numbers in python are include both integers which are whole numbers such as (0, -1, 100, ...) and float numbers that are numbers with decimal points such as (0.2, -0.34, 100.00).
 
**Strings:** they are  ordered sequence of characters in python such as ('amir', 'a', 'strings'). Keep in mind that string are immutable, which means once, they've been created, they cannot be changed or replaced. Strings support indexing and slicing.

**lists:** They are ordered sequence of objects such as ['str', 100, [1,'b']]. Unlike strings, they are mutable and allowing modifications to the element within. Same as strings, they support indexing and slicing.

**Tuples:** They are ordered immutable sequence of objects such as ('a', 100, 4.4). They are immutable, meaning their elements cannot be changed once defined. they are ideal for holding unchangeable data that remains constant throughout the program execution.

**Dictionaries:** They are unordered key:pair values. The value of a dictionary can be almost any python object. This distinctive key-value pair structure enables swift object retrieval based on keys without requiring specific index locations.
